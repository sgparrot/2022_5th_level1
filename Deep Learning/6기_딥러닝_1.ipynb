###Image data

```python
from PIL import Image
import matplotlib as plt
import numpy as np
```

```python
icebear_images=[]
icebear_resize=[] 
for image_name in range(1,5):
            image = Image.open('/content/아이스베어'+str(image_name)+'.png').convert('RGB') #경로아래에 파일이름 과 파일 type추가  #경로 설정 #.convert('RGB') #convert사용시 어떤 것으로 가져올지 L (흑백) CMYK(색의 3원색 기반)
            print(image.size) #외에도 filename(파일이름), format(형식) 등으로 파일의 정보 확인 가능
            x=np.asarray(image.resize((128,128),Image.LANCZOS)) #사이즈 조정 #필터를 줄 수도 있음, LANCZOS도 필터의 일종
            icebear_resize.append(x)
            icebear_images.append(np.asarray(image)) #원본 #resize 중 이미지 보정 등을 위한 리샘플링 옵션 설정 가능 
```
