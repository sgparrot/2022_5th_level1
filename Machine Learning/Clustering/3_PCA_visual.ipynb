```python
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()

X_digits = digits.data
y_digits = digits.target
print(X_digits.shape)
```

```python
estimator = PCA(n_components=2)
# construct covariance matrix and calculate eigenvector
# visualize in 2 axis
```

```python
X_pca = estimator.fit_transform(X_digits)
print(X_pca.shape)
# (1797, 2)
```

```python
colors = ['black', 'blue', 'purple', 'yellow', 'white', 'red', 'lime', 'cyan', 'orange', 'gray']
# 0~9 as black ~ gray
a = plt()
for i in range(len(colors)):
    
    px = X_pca[:, 0][y_digits == i]
    py = X_pca[:, 1][y_digits == i]
    plt.scatter(px, py, c=colors[i])  # assign given colors
    plt.legend(digits.target_names)
# scatter plot for dif digit. digit = i >> asign given color based on given index to the given sub data set
# target variable 0 with color[i]

plt.xlabel('first Principle Component')
plt.ylabel('second Principle Component')
plt.figure(figsize=(20,25))
```

```python
# Explained variance shows how much information can be attributed to the principle components.
print("Explained varianc: %s" % estimator.explained_variance_ratio_)
# 1  14%,   13%   .
```

```python
estimator_5 = PCA(n_components=25)
X_pca_5 = estimator_5.fit_transform(X_digits)
print(X_pca_5.shape)
```

```python
print("Explained varianc: %s" % estimator_5.explained_variance_ratio_)
```

```python
sum([i for i in estimator_5.explained_variance_ratio_])
```

```python
min_n_component = 0
for i in range(2, 100):
    estimator_loop = PCA(n_components=i)
    X_pca_loop = estimator_loop.fit_transform(X_digits)
    a = sum([i for i in estimator_loop.explained_variance_ratio_])
    if(a > 0.9):
        min_n_component = i
        break
```
